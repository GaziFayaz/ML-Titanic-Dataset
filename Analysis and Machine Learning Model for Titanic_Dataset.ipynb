Data Preprocessing

```python
titanic_data_train = titanic_data_train.drop(["Ticket", "Cabin"], axis=1)
titanic_data_test = titanic_data_test.drop(["Ticket", "Cabin"], axis=1)
```

```python
titanic_data_train.isnull().sum().sort_values(ascending=False)
```

```python
mode = titanic_data_train["Embarked"].dropna().mode()[0]
mode
```

```python
titanic_data_train["Embarked"].fillna(mode, inplace=True)
```

```python
titanic_data_test.isnull().sum().sort_values(ascending=False)
```

```python
median = titanic_data_test["Fare"].dropna().median()
median
```

```python
titanic_data_test["Fare"].fillna(median, inplace=True)
```

```python
# combine training and test set

combine = pd.concat([titanic_data_train, titanic_data_test], axis = 0).reset_index(drop = True)
combine.head()
```

```python
combine.isnull().sum().sort_values(ascending=False)
```

```python
# map male to 0 and female to 1
combine['Sex'] = combine['Sex'].map({'male': 0, 'female': 1})
```

```python
combine.head()
```

```python
age_nan_indices = list(combine[combine['Age'].isnull()].index)
len(age_nan_indices)
```

```python
for index in age_nan_indices:
    median_age = combine['Age'].median()
    predict_age = combine['Age'][(combine['SibSp'] == combine.iloc[index]['SibSp'])
                                 & (combine['Parch'] == combine.iloc[index]['Parch'])
                                 & (combine['Pclass'] == combine.iloc[index]["Pclass"])].median()
    if np.isnan(predict_age):
        combine['Age'].iloc[index] = median_age
    else:
        combine['Age'].iloc[index] = predict_age
```

```python
combine['Age'].isnull().sum()
```

Data TRansformation(Log transformation)

```python
sns.histplot(combine['Fare'], label = 'Skewness: %.2f'%(combine['Fare'].skew()))
plt.legend(loc = 'best')
plt.title('Passenger Fare Distribution')
```

```python
combine['Fare'] = combine['Fare'].map(lambda x: np.log(x) if x > 0 else 0)
```

```python
sns.histplot(combine['Fare'], label = 'Skewness: %.2f'%(combine['Fare'].skew()))
plt.legend(loc = 'best')
plt.title('Passenger Fare Distribution After Log Transformation')
```

Model the dataset

Feature Engineering (Title)

```python
combine['Title'] = [name.split(',')[1].split('.')[0].strip() for name in combine['Name']]
combine[['Name', 'Title']].head()
```

```python
combine['Title'].value_counts()
```

```python
combine['Title'].nunique()
```

```python
combine['Title'] = combine['Title'].replace(['Dr', 'Rev', 'Col', 'Major', 'Lady', 'Jonkheer', 'Don', 'Capt', 'the Countess',
                                             'Sir', 'Dona'], 'Rare')
combine['Title'] = combine['Title'].replace(['Mlle', 'Ms'], 'Miss')
combine['Title'] = combine['Title'].replace('Mme', 'Mrs')
```

```python
combine = combine.drop('Name', axis = 1)
combine.head()
```

Encoding Title

```python
combine = pd.get_dummies(combine, columns = ['Title'])
combine = pd.get_dummies(combine, columns = ['Embarked'], prefix = 'Em')
```

```python
combine.head()
```

```python
titanic_data_train = combine[:len(titanic_data_train)]
titanic_data_test = combine[len(titanic_data_train):]
titanic_data_train
```

```python
titanic_data_train = titanic_data_train.drop("PassengerId", axis=1)
titanic_data_train.head()
```

```python
titanic_data_train["Survived"] = titanic_data_train["Survived"].astype("int")
```

```python
titanic_data_train
```

```python
titanic_data_test = titanic_data_test.drop('Survived', axis = 1)
titanic_data_test.head()
```

Modeling

```python
X_train = titanic_data_train.drop('Survived', axis = 1)
Y_train = titanic_data_train['Survived']
X_test = titanic_data_test.drop('PassengerId', axis=1).copy()
print("X_train shape: ", X_train.shape)
print("Y_train shape: ", Y_train.shape)
print("X_test shape: ", X_test.shape)
# X_train
X_test
```

```python
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.svm import SVC, LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

Logistic Regression

```python
logreg = LogisticRegression(max_iter=1800)
logreg.fit(X_train, Y_train)
Y_pred = logreg.predict(X_test)
acc_log = round(logreg.score(X_train, Y_train) * 100, 2)
acc_log
```

SVC


```python
from sklearn.svm import SVC, LinearSVC
svc = SVC()
svc.fit(X_train, Y_train)
Y_pred = svc.predict(X_test)
acc_svc = round(svc.score(X_train, Y_train) * 100, 2)
acc_svc
```

K-nearest Neighbour

```python

knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train, Y_train)
Y_pred = knn.predict(X_test)
acc_knn = round(knn.score(X_train, Y_train) * 100, 2)
acc_knn
```

Stochastic GD

```python
sgd = SGDClassifier()
sgd.fit(X_train, Y_train)
Y_pred = sgd.predict(X_test)
acc_sgd = round(sgd.score(X_train, Y_train) * 100, 2)
acc_sgd
```

Training Accuracy

```python
models = pd.DataFrame({"Model": ["SVC", "KNN", "Logistic Reg.",  "Stochastic GD"],
                       "Score": [acc_svc, acc_knn, acc_log,  acc_sgd]})

models.sort_values(by="Score", ascending=False)
```

K-fold Cross Validation

```python
from sklearn.model_selection import cross_val_score
```

```python
classifiers = []
classifiers.append(LogisticRegression(max_iter=1500))
classifiers.append(SVC())
classifiers.append(KNeighborsClassifier(n_neighbors=5))

classifiers.append(SGDClassifier())

len(classifiers)
```

```python
cv_results = []
for classifier in classifiers:
    cv_results.append(cross_val_score(classifier, X_train, Y_train, scoring="accuracy", cv=10))
```

```python
cv_mean = []
cv_std = []
for cv_result in cv_results:
    cv_mean.append(cv_result.mean())
    cv_std.append(cv_result.std())
```

```python
cv_res = pd.DataFrame({'Cross Val Mean': cv_mean, 'Cross Val Std': cv_std, 'Algo': ['Logistic Regression', 'SVC', 'KNN', 'StochasticGD']})
cv_res.sort_values(by = 'Cross Val Mean', ascending = False)
```
